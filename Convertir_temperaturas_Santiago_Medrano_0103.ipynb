## 1. Fórmulas de Conversión

Para convertir entre escalas, utilizamos las siguientes relaciones matemáticas:

* **De Celsius a Fahrenheit:** Multiplicamos los grados Celsius por $1.8$ (o $9/5$) y sumamos $32$.
    $$°F = (°C \times 1.8) + 32$$
* **De Fahrenheit a Celsius:** Restamos $32$ a los grados Fahrenheit y dividimos el resultado entre $1.8$.
    $$°C = \frac{°F - 32}{1.8}$$

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo A: Convertir 25°C a °F (Temperatura ambiente)
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$
3.  **Multiplicar:** $25 \times 1.8 = 45$
4.  **Sumar 32:** $45 + 32 = 77$
* **Resultado:** $25°C$ equivalen a **$77°F$**.

### Ejemplo B: Convertir 98.6°F a °C (Temperatura corporal)
1.  **Identificar la fórmula:** $°C = (98.6 - 32) / 1.8$
2.  **Restar 32:** $98.6 - 32 = 66.6$
3.  **Dividir entre 1.8:** $66.6 / 1.8 = 37$
* **Resultado:** $98.6°F$ equivalen a **$37°C$**.

## 3. Código en Python

Este programa sigue una estructura modular, utilizando funciones para separar la lógica de los cálculos de la interacción con el usuario.

```python
def convertirCF(celsius):
    """Convierte de Celsius a Fahrenheit."""
    fahrenheit = (celsius * 1.8) + 32
    return fahrenheit

def convertirFC(fahrenheit):
    """Convierte de Fahrenheit a Celsius."""
    celsius = (fahrenheit - 32) / 1.8
    return celsius

def menu():
    """Controla el flujo principal del programa."""
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir °C a °F")
        print("2. Convertir °F a °C")
        print("3. Salir")
        
        opcion = input("Seleccione una opción (1-3): ")

        if opcion == '1':
            try:
                c = float(input("Ingrese los grados Celsius: "))
                resultado = convertirCF(c)
                print(f"Resultado: {c}°C es igual a {resultado:.2f}°F")
            except ValueError:
                print("Error: Por favor, ingrese un número válido.")

        elif opcion == '2':
            try:
                f = float(input("Ingrese los grados Fahrenheit: "))
                resultado = convertirFC(f)
                print(f"Resultado: {f}°F es igual a {resultado:.2f}°C")
            except ValueError:
                print("Error: Por favor, ingrese un número válido.")

        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            break
        
        else:
            print("Opción no válida. Intente de nuevo.")

# Punto de entrada al programa
if __name__ == "__main__":
    menu()
```

### Explicación del código:
* **`while True`**: Crea un ciclo infinito para que el menú se repita hasta que tú decidas salir.
* **`float()`**: Permite que el usuario ingrese números decimales (como 36.5).
* **`try/except`**: Es una buena práctica para evitar que el programa se cierre si el usuario escribe letras en lugar de números.
* **`f"{valor:.2f}"`**: Se usa para mostrar solo 2 decimales en el resultado final, haciendo que se vea más limpio.

````python
def convertirCF(celsius):
    """Convierte de Celsius a Fahrenheit."""
    fahrenheit = (celsius * 1.8) + 32
    return fahrenheit

def convertirFC(fahrenheit):
    """Convierte de Fahrenheit a Celsius."""
    celsius = (fahrenheit - 32) / 1.8
    return celsius

def menu():
    """Controla el flujo principal del programa."""
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir °C a °F")
        print("2. Convertir °F a °C")
        print("3. Salir")

        opcion = input("Seleccione una opción (1-3): ")

        if opcion == '1':
            try:
                c = float(input("Ingrese los grados Celsius: "))
                resultado = convertirCF(c)
                print(f"Resultado: {c}°C es igual a {resultado:.2f}°F")
            except ValueError:
                print("Error: Por favor, ingrese un número válido.")

        elif opcion == '2':
            try:
                f = float(input("Ingrese los grados Fahrenheit: "))
                resultado = convertirFC(f)
                print(f"Resultado: {f}°F es igual a {resultado:.2f}°C")
            except ValueError:
                print("Error: Por favor, ingrese un número válido.")

        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            print("Santiago Medrano NL = 37")
            break

        else:
            print("Opción no válida. Intente de nuevo.")

# Punto de entrada al programa
if __name__ == "__main__":
    menu()
````
