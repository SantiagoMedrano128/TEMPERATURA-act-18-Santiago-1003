## 1. Fórmulas de Conversión

Para convertir entre estas dos escalas, utilizamos las siguientes relaciones matemáticas basadas en los puntos de congelación (0°C / 32°F) y ebullición (100°C / 212°F) del agua:

* **De Celsius a Fahrenheit:**
    $$°F = (°C \times \frac{9}{5}) + 32$$
    *(También se puede usar $1.8$ en lugar de $9/5$)*

* **De Fahrenheit a Celsius:**
    $$°C = \frac{(°F - 32) \times 5}{9}$$

---

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo 1: Convertir 25°C a Fahrenheit
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$
3.  **Multiplicar:** $25 \times 1.8 = 45$
4.  **Sumar 32:** $45 + 32 = 77$
**Resultado:** 25°C equivalen a **77°F**.

### Ejemplo 2: Convertir 95°F a Celsius
1.  **Identificar la fórmula:** $°C = (95 - 32) \times \frac{5}{9}$
2.  **Restar 32:** $95 - 32 = 63$
3.  **Multiplicar por 5:** $63 \times 5 = 315$
4.  **Dividir entre 9:** $315 / 9 = 35$
**Resultado:** 95°F equivalen a **35°C**.

## 3. Código en Python

```python
# Programa de conversión de temperaturas

def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    c = float(input("\nIngresa los grados Celsius: "))
    f = (c * 9/5) + 32
    print(f"--- {c}°C equivalen a {f:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    f = float(input("\nIngresa los grados Fahrenheit: "))
    c = (f - 32) * 5/9
    print(f"--- {f}°F equivalen a {c:.2f}°C ---")

# Bucle principal del programa

    elif opc == '2':
        convertirFC()
```

### Explicación breve del código:
* **`while True`**: Crea un bucle infinito que permite realizar múltiples conversiones sin cerrar el programa.
* **`float()`**: Se usa para que el usuario pueda ingresar decimales (como 36.5°C).
* **`break`**: Es la instrucción que rompe el bucle y finaliza el programa cuando eliges la opción 3.
* **`f-strings`**: Se usan en los `print` para insertar las variables y formatear el resultado a dos decimales (`:.2f`).

````python
# Programa de conversión de temperaturas
# Autor: Santiago Merano NL = 37

def convertirCF():
    """Convierte de Celsius a Fahrenheit"""
    c = float(input("\nIngresa los grados Celsius: "))
    f = (c * 9/5) + 32
    print(f"--- {c}°C equivalen a {f:.2f}°F ---")

def convertirFC():
    """Convierte de Fahrenheit a Celsius"""
    f = float(input("\nIngresa los grados Fahrenheit: "))
    c = (f - 32) * 5/9
    print(f"--- {f}°F equivalen a {c:.2f}°C ---")

def menu():
    """Muestra el menú de opciones"""
    print("\n==============================")
    print("      MENÚ DE CONVERSIONES    ")
    print("==============================")
    print("1.- Convertir °C a °F")
    print("2.- Convertir °F a °C")
    print("3.- Salir")
    opcion = input("Selecciona una opción: ")
    return opcion

# Bucle principal del programa
while True:
    opc = menu()

    if opc == '1':
        convertirCF()
    elif opc == '2':
        convertirFC()
    elif opc == '3':
        print("\n¡Gracias por usar el conversor!")
        print("Programa realizado por Santiago Medrano NL = 37")
        break
    else:
        print("\nOpción no válida, intenta de nuevo.")
````
